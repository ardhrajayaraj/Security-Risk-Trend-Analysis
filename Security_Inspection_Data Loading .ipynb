### Objective  
The objective of this project is to analyze security inspection data across healthcare facilities to evaluate inspection performance and identify risk and non-compliance trends. The study focuses on understanding variations across facilities, departments, and inspection types using exploratory data analysis in Python. By analyzing patterns in scores, incidents, and follow-up actions, the project aims to generate data-driven insights to support improved security quality and risk management.  

### Problem Definition
Healthcare facilities conduct a large number of security inspections to ensure safety, compliance, and risk mitigation; however, inspection data is often underutilized and not systematically analyzed to identify performance gaps and risk patterns. The absence of data-driven analysis makes it difficult to understand trends in non-compliance, inspection effectiveness, and follow-up actions across facilities and departments. This project addresses this problem by applying Python-based exploratory data analysis to security inspection data to uncover meaningful patterns, correlations, and risk indicators that can support informed decision-making and continuous improvement in security quality management. 

### Dataset Description  
The dataset contains inspection-level information including operational attributes, time-based variables, and categorical inspection classifications.

### Data Quality Checks

```python

# Missing Values Summary
print("Missing Values Summary:")
missing_values = df.isnull().sum()
print(missing_values)
print("="*70)

if missing_values.sum() > 0:
    print("\nColumns with missing values:")
    print(missing_values[missing_values > 0])
else:
    print("\nNo missing values found in the dataset.")

# Duplicate Rows Check
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

if duplicates > 0:
    print(f"Warning: {duplicates} duplicate row(s) detected!")
else:
    print("No duplicate rows found in the dataset.")
```
